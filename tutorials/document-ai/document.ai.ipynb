```python
# Import Dependencies
from hammadml.data import Database

database = Database()
```

```python
# Will search through a directory for pdfs, docs, and txt files and load them into the database.
# The documents are parsed into paragraphs and sentences.
database.load_docs(dir="docs")
results = database.search(query="Barry Benson")

for result in results:
    print(result['content'][:500])
```
