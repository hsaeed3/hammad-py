```python
# Initialize

from hammadpy import Core
core = Core()
```

```python
core.say("Hello!")
core.say("Colors can be different", color="blue")
core.say("You can use rgb too", color="rgb(255, 0, 0)")
core.say("You can change the background", bg="white", color="black")
core.say("rgb works here too of course", bg="rgb(255, 0, 0)")
core.say("or underline...", underline=True)
core.say("OR BOLD", bold=True)
core.say("OR ITALIC", italic=True)
```
